3. Faça um Programa que leia 40 notas, mostre as notas e a média na tela.

4. Faça um Programa que leia 20 números inteiros e armazene-os numa lista. Armazene os números pares na lista PAR e os números IMPARES na lista impar. Imprima as três listas.

```python
totalLista = 0
par = []
impar = []

while totalLista < 20:
    try:
        num = int(input('Digite um número inteiro: '))
    except:
        print('Não é um número e inteiro')
    else:
        if num >= 0:
            if num%2 == 0:
                par.append(num)
            else:
                impar.append(num)
            totalLista += 1
        else:
            print('O número deve ser maior ou igual a zero')

print(f'Pares: {par}\nImpares: {impar}')
```

5. Faça um Programa que peça as quatro notas de 10 alunos, calcule e armazene numa lista a média de cada aluno, imprima o número de alunos com média maior ou igual a 7.0.

```python
listaMedias = []
mediasAltas = 0


for c in range(0, 10):
    somatorioNotas = 0
    notasAlunos = []
    media = 0
    totalNotas = 0

    while totalNotas < 4:
        try:
            nota = float(input(f'Digite a {totalNotas+1}° nota do {c+1}° aluno: ').replace(',', '.'))
        except:
            print('Não é um número')
        else:
            if nota >= 0 and nota <= 10:
                notasAlunos.append(nota)
                totalNotas += 1
            else:
                print('Esse número deve estar entre 0 e 10')

    for n in notasAlunos:
        somatorioNotas += n

    media = somatorioNotas/totalNotas
    
    if media >= 7:
        mediasAltas += 1
    
    listaMedias.append(media)

print(f'O número de alunos com média maior ou igual a 7 foi: {mediasAltas}')
```

7. Faça um Programa que peça a idade e a altura de 10 pessoas, armazene cada informação na sua respectiva lista. Imprima a idade da pessoa que possui maior altura.

```python
totalPessoas = 0
alturas = []
idades = []

while totalPessoas < 10:
    try:
        idade = int(input(f'Digite a idade do {totalPessoas+1}° participante: '))
        altura = float(input(f'Digite a altura (em metros) do {totalPessoas+1}° participante: ').replace(',', '.'))
    except:
        print('Escolha inválida')
    else:
        if idade > 0 and altura > 0:
            idades.append(idade)
            alturas.append(altura)
            totalPessoas += 1
        else:
            print('Não se pode atribuir valores negativos')

maior = max(alturas)
idadeMaior = idades[alturas.index(maior)]

print(f'A pessoa com maior altura tem {idadeMaior} anos e tem {maior}m de altura')
```

11. Foram anotadas as idades e alturas de 30 alunos. Faça um Programa que determine quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos.

```python
media = 0
somatorio = 0
totalPessoas = 0
alturas = []
idades = []
alunosPequenos = 0

while totalPessoas < 30:
    try:
        idade = int(input(f'Digite a idade do {totalPessoas+1}° aluno: '))
        altura = float(input(f'Digite a altura (em metros) do {totalPessoas+1}° aluno: ').replace(',', '.'))
    except:
        print('Escolha inválida')
    else:
        if idade > 0 and altura > 0:
            idades.append(idade)
            alturas.append(altura)
            totalPessoas += 1
        else:
            print('Não se pode atribuir valores negativos')

for tamanho in alturas:
    somatorio += tamanho

media = somatorio/totalPessoas

for i in range(0, len(alturas)):
    if idades[i] > 13 and alturas[i] < media:
        alunosPequenos += 1

print(f'A quantidade de alunos com mais de 13 anos e com altura menor que a média geral é: {alunosPequenos}')
```

13. Faça um programa que receba a temperatura média de cada mês do ano e armazene-as em uma lista. Após isto, calcule a média anual das temperaturas e mostre todas as temperaturas acima da média anual, e em que mês elas ocorreram (mostrar o mês por extenso: 1 – Janeiro, 2 – Fevereiro, . . . ).

```python
mesesAnalisados = 0
tempMeses = []
somatorio = 0
mediaAnual = 0
meses = {
    1: "Janeiro",
    2: "Fevereiro",
    3: "Março",
    4: "Abril",
    5: "Maio",
    6: "Junho",
    7: "Julho",
    8: "Agosto",
    9: "Setembro",
    10: "Outubro",
    11: "Novembro",
    12: "Dezembro"
}

while mesesAnalisados < 12:
    try:
        temp = float(input(f'Digite a média das temperaturas no mês de {meses[mesesAnalisados+1]} em °C: ').replace(',', '.'))
    except:
        print('O valor indicado não é um número')
    else:
        somatorio += temp
        tempMeses.append(temp)
        mesesAnalisados += 1

mediaAnual = somatorio/mesesAnalisados

print(f'A média anual foi de {mediaAnual:.2f} °C')

for posLista in range(len(tempMeses)):
    if tempMeses[posLista] > mediaAnual:
        print(f'A temperatura do mês de {meses[posLista+1]} de {tempMeses[posLista]}, sendo essa maior que a média anual')
```

14. Faça um programa que leia um número indeterminado de valores, correspondentes a notas, encerrando a entrada de dados quando for informado um valor igual a -1 (que não deve ser armazenado). Após esta entrada de dados, faça:
<pre>
a. Mostre a quantidade de valores que foram lidos;
b. Exiba todos os valores na ordem em que foram informados
c. Exiba todos os valores na ordem inversa à que foram informados
d. Calcule e mostre a soma dos valores;
e. Calcule e mostre a média dos valores;
f. Calcule e mostre a quantidade de valores acima da média calculada;
g. Calcule e mostre a quantidade de valores abaixo de sete;
h. Encerre o programa com uma mensagem.
</pre>

```python
permitirLeitura = True
notas = []
notasAltas = 0
notasBaixas = 0
somatorio = 0
quantidadeLista = 0
media = 0

while permitirLeitura:
    try:
        nota = float(input(f'Digite a {quantidadeLista+1}ª nota (-1 para sair): ').replace(',', '.'))
    except:
        print('O valor não é um número')
    else:
        if nota == -1:
            permitirLeitura = False

        if nota >= 0 and nota <= 10:
            somatorio += nota
            quantidadeLista += 1
            notas.append(nota)

        elif nota != -1:
            print('Só são permitidos os valores entre 0 e 10 e também o -1 caso queira encerrar o programa')

if quantidadeLista > 0:
    media = somatorio/quantidadeLista

    for notaAnalisada in notas:
        if notaAnalisada > media:
            notasAltas += 1

        if notaAnalisada < 7:
            notasBaixas += 1

    print(f"""
A quantidade de valores lidos foi: {len(notas)}
A lista (na ordem informada) foi: {notas}
A lista (na ordem inversa) foi: {notas[::-1]}
O somatório dos valores foi: {somatorio}
A média dos valores foi: {media}
A quantidade de valores acima da média foi: {notasAltas}
A quantidade de valores abaixo de 7 foi: {notasBaixas}
    """)
    print('Muito obrigado!')
else:
    print('É preciso informar mais de um valor válido')
```

O programa deverá ler os valores até ser informado o valor 0, que encerra a entrada dos dados. Não deverão ser aceitos valores além dos válidos para o programa (0 a 6). Os valores referentes a cada uma das opções devem ser armazenados numa lista. Após os dados terem sido completamente informados, o programa deverá calcular o percentual de cada um dos concorrentes e informar o vencedor da enquete. 16. Escreva um programa em Python para encontrar o segundo maior elemento de uma lista com 20 números inteiros.

```python
totalLista = 0
listaInt = []
listaSemRepticoes = []
listaOrdenada = []

while totalLista < 10:
    try:
        num = int(input('Digite um número inteiro para armazenar na lista: '))
    except:
        print('Não é um número e inteiro')
    else:
        listaInt.append(num)
        totalLista += 1

for i in listaInt:
    if not i in listaSemRepticoes:
        listaSemRepticoes.append(i)

listaOrdenada = listaSemRepticoes.copy()
listaOrdenada.sort()

print(f'O segundo maior número é: {listaOrdenada[-2]}')
```

17. Escreva um programa em Python para converter um número inteiro em binário de acordo com a representação de grandeza com sinal (sinal e magnitude). O programa deve receber um número inteiro e produzir como saída uma lista com os bits do número convertido (um bit para cada posição da lista). Além disso deve ser feita a verificação se o número pode ser representado, considere uma representação com 8 bits (um para o sinal e 7 para a magnitude).

```python
binario = [0, 0, 0, 0, 0, 0, 0, 0]
escolhaNaoConcluida = True

while escolhaNaoConcluida:
    try:
        numDecimal = int(input('Informe um número inteiro para converter para binário: '))
    except:
        print('Não é um número e inteiro')
    else:
        if numDecimal > -128 and numDecimal < 128:
            if numDecimal < 0:
                binario[0] = 1
                numDecimal *= -1
                
            for numBin in range(len(binario)-1, 1, -1):
                bit = numDecimal%2
                numDecimal //= 2
                binario[numBin] = bit

            escolhaNaoConcluida = False
        else:
            print('O número deve estar entre -128 e 128')
            
print(f'A lista dos bits do número convertido é: {binario}')
```

18. Leia uma matriz 4 x 4, conte e escreva quantos valores maiores que 10 ela possui.

```python
matriz = [
        [],
        [],
        [],
        []
        ]
linha = 1
quantidadeGrande = 0

for i in matriz:
    for j in range(0, len(matriz)):
        manterColuna = True
        
        while manterColuna:
            try:
                num = float(input(f'Digite o valor da matriz A{linha}{j+1} ').replace(',', '.'))
            except:
                print('Esse valor não é um número')
            else:
                i.append(num)
                manterColuna = False
                if num > 10:
                    quantidadeGrande += 1
    linha += 1

print(f'A matriz possui {quantidadeGrande} números maiores que 10')
```

20. Leia uma matriz 5 x 5. Leia também um valor X. O programa deverá fazer uma busca desse valor na matriz e, ao final, escrever a localizacão (linha e coluna) ou uma mensagem de “não encontrado”.

```python
matriz = [
        [],
        [],
        [],
        [],
        [],
        ]
escolhaNaoEncontrada = True
encontrado = False
linha = 1

for i in matriz:
    for j in range(0, len(matriz)):
        manterColuna = True
        
        while manterColuna:
            try:
                num = float(input(f'Digite o valor da matriz A{linha}{j+1} ').replace(',', '.'))
            except:
                print('Esse valor não é um número')
            else:
                i.append(num)
                manterColuna = False
    linha += 1

while escolhaNaoEncontrada:
    try:
        valorX = float(input('Digite um valor para ser procurado na matriz: '))
    except:
        print('O valor não é um número')
    else:
        for i in matriz:
            if valorX in i:
                escolhaNaoEncontrada = False
                encontrado = True
                
        escolhaNaoEncontrada = False

if encontrado:
    print(f'O valor {valorX} está na matriz!')
else:
    print(f'O valor {valorX} não está na matriz')
```
